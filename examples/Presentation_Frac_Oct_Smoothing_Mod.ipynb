# Fractional Octave Smoothing

```python
from pyfar.dsp import fractional_octave_smoothing as fs
from pyfar.dsp import dsp as dsp
from pyfar import Signal
import numpy as np

import pyfar.plot as plt
```

```python
# Source signal:
np.random.seed(1)
shape = (1, 2205)
src_magn = np.ones(shape)
src_phase = np.random.random(shape)
f_s = 44100
fft_norm = 'none'
white_noise = Signal(src_magn*np.exp(1j*2*np.pi*src_phase), f_s, domain='freq', fft_norm=fft_norm)
```

```python
# Smoothing objects
win_width = 3
phase_type = fs.PhaseType.ORIGINAL
# Zero padding
paddig_type = fs.PaddingType.ZERO
output_zero_padding = dsp.fract_oct_smooth(white_noise, win_width,
                                           n_bins=None,
                                           phase_type=phase_type,
                                           padding_type=paddig_type)
# Edge padding
paddig_type = fs.PaddingType.EDGE
output_edge_padding = dsp.fract_oct_smooth(white_noise, win_width,
                                           n_bins=None,
                                           phase_type=phase_type,
                                           padding_type=paddig_type)
# Mean padding
paddig_type = fs.PaddingType.MEAN
output_mean_padding = dsp.fract_oct_smooth(white_noise, win_width,
                                           n_bins=None,
                                           phase_type=phase_type,
                                           padding_type=paddig_type)
```

```python
plt.line.time(white_noise, style='dark')
plt.line.freq(output_zero_padding, style='dark')
plt.line.freq(output_edge_padding, style='dark')
plt.line.freq(output_edge_padding, style='dark')
```
